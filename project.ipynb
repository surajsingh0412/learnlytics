```python
###   1.-- How many students on same address??
ax=sns.countplot(x='address',data=df)
ax.set_title("Student Same Address")
for bars in ax.containers:  ### For counting  
    ax.bar_label(bars)
```

```python
df.rename(columns={'sex':'Gender'},inplace=True)
```

```python
###   2.Based on gender--------(Total gender )
ax=sns.countplot(x='Gender',data=df)
ax.set_title('Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
###   3 .Based on gender with location wise
ax=sns.countplot(data=df,x='address',hue='Gender')
ax.set_title('Gender Distribution by address')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#### 4.  Gender Distribuation in U address
df_U = df[df["address"] == "U"]
ax = sns.countplot(data=df_U, x="Gender")
ax.set_title("Gender Distribution in U address", fontsize=14)
ax.set_xlabel("Gender", fontsize=12)
ax.set_ylabel("Count", fontsize=12)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
###  5. Gender distribuation in R address
df_R=df[df["address"]=="R"]
ax=sns.countplot(data=df_R,x="Gender")
ax.set_title("Gender Distribution in R address", fontsize=12)
ax.set_xlabel("Gender",fontsize=10)
ax.set_ylabel("count",fontsize=10)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#####   6.-- Job and jobless based(house wife)
Mother_job_counts=df['Mother_job'].value_counts()
ax=sns.barplot(x=Mother_job_counts.index,y=Mother_job_counts.values,color='red', edgecolor="black",alpha=0.9)
plt.title("Count of Mother job", fontsize=14)
plt.xlabel("Mother's Jobs",fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.xticks(rotation=45)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
### 7. Distribution of Health Status Among Students
health_counts = df['health_status'].value_counts().sort_index()
ax=plt.bar(health_counts.index, health_counts.values, color='skyblue')
plt.title('Distribution of Health Status Among Students')
plt.xlabel('Health Status (1=Very Bad, 5=Very Good)')
plt.ylabel('Number of Students')
bars = plt.gca().patches
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height}', 
             ha='center', va='bottom', fontsize=10, fontweight='bold')
# plt.xticks(range(1, 6))  
plt.show()
```

```python
### 7.1- Distribution of Health Status Among Students
# health_counts = df['health_status'].value_counts().sort_index()
sns.histplot(df['health_status'], bins=5, kde=True, color='blue')
plt.title('Distribution of Health Status Among Students')
plt.xlabel('Health Status (1=Very Bad, 5=Very Good)')
plt.ylabel('Number of Students')
bars = plt.gca().patches
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height}', 
             ha='center', va='bottom', fontsize=10, fontweight='bold')
# plt.xticks(range(1, 6))  
plt.show()
```

```python
###  8. How many student take extra_paid_class
extra_paid_count = df['extra_paid_class'].value_counts()
# Plot a pie chart for the distribution
plt.figure(figsize=(6, 6))
extra_paid_count.plot(kind='pie', autopct='%1.1f%%', colors=['blue','red'],fontsize=16, startangle=90)
plt.title('Proportion of Students Taking Extra Paid Classes',fontsize=16)
plt.ylabel('')  
plt.show()
```

```python
###  9. Interested in hire Education
hire_Education = df['Interested_in_higher_edu'].value_counts()
# Plot a pie chart for the distribution
plt.figure(figsize=(6, 6))
ax=hire_Education.plot(kind='bar')
plt.title('Interested in Hire Education',fontsize=16)
plt.ylabel('Count') 
plt.xlabel(' ')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=0)
plt.show()
```

```python
###  9.1.  How many student are Interested in hire Education.
hire_Education = df['Interested_in_higher_edu'].value_counts()
# Plot a pie chart for the distribution
plt.figure(figsize=(6, 6))
ax=hire_Education.plot(kind='pie',autopct="%1.1f%%",colors=["skyblue","green"],fontsize=16, startangle=90)
plt.title('Interested in Hire Education',fontsize=16)
plt.ylabel('')  
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=0)
plt.show()
```

```python
### 10. Extra curriculam activity (freetime_after_school)
extra=df["extra_curr_activities"].value_counts()
ax=extra.plot(kind='bar',fontsize=16)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel("Extra curriculam activity")
plt.ylabel("Count")
plt.title("Extra curriculam activity (freetime_after_school)")
plt.xticks(rotation=360)
plt.show()
```
